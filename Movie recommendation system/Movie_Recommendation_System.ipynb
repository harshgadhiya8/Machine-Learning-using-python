```python
from pandas.core.ops import methods
movie_correlations = user_matrix.corr(method='pearson',min_periods=80)
```

```python
movie_correlations
```

```python
myratings = pd.read_csv('/content/drive/MyDrive/Data/Movie Review/My_Ratings.csv')
```

```python
myratings
```

```python
similar_movies = pd.Series()

for i in range(0,2):
  list1 = movie_correlations[myratings['Movie Name'][i]].dropna()
  list1 = list1.map(lambda x :x*myratings['Ratings'][i])
  similar_movies = similar_movies.append(list1)
```

```python
similar_movies.sort_values(inplace=True,ascending=False)
print(similar_movies)
```
